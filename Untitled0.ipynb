```python
import pandas as pd

# Dosyayı yükleyip veri çerçevesine aktarma
file_path = "/content/udemy_courses.csv"
data = pd.read_csv(file_path)

# price sütunundaki değerlerin dağılımını inceleyelim
print("Price sütunundaki değerlerin dağılımı:")
print(data['price'].describe())

# price sütununda NaN değer olup olmadığını kontrol etme
print("\nEksik Veri Sayıları (price için):")
print(data['price'].isnull().sum())

# price sütununda NaN varsa ortalama ile doldurmak
data['price'] = data['price'].fillna(data['price'].mean())

# price sütunundaki aralıklara göre price_range oluşturma
price_bins = [0, 50, 100, 150, 200, 250, 1000]  # Fiyat aralıkları
price_labels = ['0-50', '51-100', '101-150', '151-200', '201-250', '250+']  # Etiketler
data['price_range'] = pd.cut(data['price'], bins=price_bins, labels=price_labels, right=False)

# price_range sütununda eksik veri olup olmadığını kontrol etme
print("\nEksik Veri Sayıları (price_range için):")
print(data['price_range'].isnull().sum())



# Veri setinin ilk birkaç satırını ve genel bilgilerini görüntüleme
print("\nVeri Setinin İlk 5 Satırı:")
print(data.head())

# Temizlenmiş veri setini kaydetme
cleaned_file_path = "/content/cleaned_udemy_courses.csv"
data.to_csv(cleaned_file_path, index=False)
print(f"\nTemizlenmiş veri seti '{cleaned_file_path}' olarak kaydedildi.")
```

```python
# Gerekli kütüphaneler
from collections import Counter  # Kelime frekansını hesaplamak için kullanılır
import matplotlib.pyplot as plt  # Görselleştirme için kullanılır
import seaborn as sns  # Daha estetik grafikler için Matplotlib üzerine inşa edilmiş bir kütüphane

# 1. Kurs başlıklarındaki tüm kelimeleri birleştirerek analiz etmek
# Tüm başlıkları birleştir, küçük harfe çevir ve kelimelere ayır
all_titles = ' '.join(data['course_title']).lower()
words = all_titles.split()

# 2. Kelime frekansını hesaplama
# En sık geçen kelimeleri bulmak için Counter kullanılır
word_counts = Counter(words)  # Tüm kelimelerin frekansını hesaplar
top_words = word_counts.most_common(10)  # En çok geçen 10 kelimeyi alır

# 3. Top 10 kelimenin geçtiği kurs sayısını hesaplama
# Her kelimenin kurs başlıklarında kaç kez geçtiğini kontrol eder
word_count_data = []  # Sonuçları depolamak için boş liste

for word, _ in top_words:
    # Belirli bir kelimenin kurs başlıklarında kaç kez geçtiğini hesapla
    word_count = data['course_title'].str.lower().str.contains(word).sum()
    word_count_data.append((word, word_count))

# 4. Kelime ve kurs sayısını veri çerçevesine aktarma
# Daha kolay görselleştirme için bir DataFrame oluşturuyoruz
top_words_df = pd.DataFrame(word_count_data, columns=['Word', 'Course Count'])

# 5. Sonuçları çubuk grafikte görselleştirme
plt.figure(figsize=(10, 6))  # Grafik boyutunu ayarlama
sns.barplot(data=top_words_df, x='Word', y='Course Count', palette='plasma')  # Çubuk grafiği oluşturma
plt.title('En Popüler Kelimeler ve Bu Kelimeleri İçeren Kurs Sayısı')  # Grafik başlığı
plt.xlabel('Kelime')  # X ekseni etiketi
plt.ylabel('Kurs Sayısı')  # Y ekseni etiketi
plt.xticks(rotation=45)  # X eksenindeki kelimeleri yatay yerine eğimli gösterme
plt.tight_layout()  # Grafik kenarlarının düzenli görünmesini sağlar
plt.show()  # Grafiği gösterme
```
